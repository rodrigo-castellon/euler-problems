```python
def expand(n):
    i = 0
    num = math.pi
    u = [num]
    a = []
    while i < n:
        a.append(math.floor(u[-1]))
        u.append(1/(u[-1]-a[-1]))
        i += 1
    return a
```

```python
def isprime(n):
	if n <= 1:
		return False
	if n <= 3:
		return True
	if n % 2 == 0 or n % 3 == 0:
		return False
	i = 5
	while i*i <= n:
		if n % i == 0 or n % (i+2) == 0:
			return False
		i += 6
	return True
```

```python
def factorize(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i == 0 and isprime(i):
            factors.append(i)
            if i*i != n and isprime(n//i):
                factors.append(n//i)
            n = n//i
        i = i + 1
    return sorted(factors)
```

```python
def gcd(a,b):
    while b:
        a,b = b, a%b
    return a
```

```python
def simplify(frac):
    start = time.time()
    # frac is 2-length list
    dem = gcd(frac[0],frac[1])
    return frac[0]//dem, frac[1]//dem
```

```python
def find_expansion(expansion_num=3):
    start = time.time()
    i = 0
    top = 5
    bottom = 2
    i += 1
    while i < expansion_num:
        if expansion_num - i == 1:
            top_new = top+bottom
            bottom_new = top
            top = top_new
            bottom = bottom_new
        else:
            top_new = 2*top+bottom
            bottom_new = top
            top = top_new
            bottom = bottom_new
        i += 1
    print('expansion {} took {:.2}s'.format((str(top) + '/' + str(bottom)), (time.time()-start)))
    return simplify([top, bottom])
```

```python
def does_expansion_satisfy(expansion_num):
    expansion = find_expansion(expansion_num)
    return True if len(str(expansion[0])) > len(str(expansion[1])) else False
```

```python
satisfies_array = []
for i in range(1,1001):
    print(i)
    #print(find_expansion(i))
    satisfies_array.append(does_expansion_satisfy(i))
```
